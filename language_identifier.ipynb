```python
import string
import re
import codecs
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import feature_extraction 
from sklearn import linear_model 
from sklearn import pipeline
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
with open("mtop/en/train.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Clean or preprocess lines if needed
cleaned_lines = [line.strip() for line in lines]

# Create a DataFrame
eng_df = pd.DataFrame({"English": cleaned_lines})
```

```python
with open("mtop/es/train.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Clean or preprocess lines if needed
cleaned_lines = [line.strip() for line in lines]

# Create a DataFrame
es_df = pd.DataFrame({"Spanish": cleaned_lines})
es_df.head()
```

```python
with open("mtop/hi/train.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Clean or preprocess lines if needed
cleaned_lines = [line.strip() for line in lines]

# Create a DataFrame
hi_df = pd.DataFrame({"Hindi": cleaned_lines})
hi_df.head()
```

```python
with open("mtop/th/train.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Clean or preprocess lines if needed
cleaned_lines = [line.strip() for line in lines]

# Create a DataFrame
th_df = pd.DataFrame({"Thai": cleaned_lines})
th_df.head()
```

```python
for char in string.punctuation:
    print (char, end=" ")
translate_table = dict((ord(char), None) for char in string.punctuation)
```

```python
data_eng = []
lang_eng = []
```

```python
for i, line in eng_df.iterrows():
    line = line["English"]
    if len(line) != 0:
        line = line.lower()
        line = re.sub(r"\d+", "", line)
        line = line.translate(translate_table)
        data_eng.append(line)
        lang_eng.append("English")
```

```python
data_eng
```

```python
lang_de = []
data_de = []

for i, line in de_df.iterrows():
    line = line["German"]
    if len(line) != 0:
        line = line.lower()
        line = re.sub(r"\d+", "", line)
        line = line.translate(translate_table)
        data_de.append(line)
        lang_de.append("German")
```

```python
lang_es = []
data_es = []

for i, line in es_df.iterrows():
    line = line["Spanish"]
    if len(line) != 0:
        line = line.lower()
        line = re.sub(r"\d+", "", line)
        line = line.translate(translate_table)
        data_es.append(line)
        lang_es.append("Spanish")
```

```python
data_es
```

```python
lang_hi = []
data_hi = []

for i, line in hi_df.iterrows():
    line = line["Hindi"]
    if len(line) != 0:
        line = line.lower()
        line = re.sub(r"\d+", "", line)
        line = re.sub(r"[a-zA-Z]", "", line)
        line = line.translate(translate_table)
        data_hi.append(line)
        lang_hi.append("Hindi")
```

```python
lang_th = []
data_th = []

for i, line in th_df.iterrows():
    line = line["Thai"]
    if len(line) != 0:
        line = line.lower()
        line = re.sub(r"\d+", "", line)
        line = re.sub(r"[a-zA-Z]", "", line)
        line = line.translate(translate_table)
        data_th.append(line)
        lang_th.append("Thai")
```

```python
data_th
```

```python
df = pd.DataFrame({"Text": data_eng + data_de + data_es + data_hi + data_th, "Language":lang_eng + lang_de + lang_es + lang_hi + lang_th})
```

```python
df.head()
```

```python
df.shape
```

```python
X, y = df.iloc[:,0], df.iloc[:,1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train
```

# Vectoriser and Model fitting Pipeline

```python
vectorizer = feature_extraction.text.TfidfVectorizer(ngram_range = (1, 3), analyzer = 'char')
```

```python
pipe_lr_r13 = pipeline.Pipeline([
    ('vectorizer', vectorizer),
    ('clf', linear_model.LogisticRegression())
])
```

```python
# Model fitting
pipe_lr_r13.fit(X_train, y_train)
```

```python
y_predicted = pipe_lr_r13.predict(X_test)
```

```python
acc = (metrics.accuracy_score(y_test, y_predicted))*100
acc
```

```python
matrix = metrics.confusion_matrix(y_test, y_predicted)
print('confusion matrix : \n', matrix)
```

```python
import pickle

lrFile = open('LRModel.pckl', 'wb')
pickle.dump(pipe_lr_r13, lrFile)
lrFile.close()
```

# Method Definition to call trained Model and make Predictions

```python
def lang_detect(text):
    import string
    import re
    import numpy as np
    import pickle
    translate_table = dict((ord(char), None) for char in string.punctuation)

    global lrLangDetectModel
    lrLangDetectFile = open('LRModel.pckl','rb')
    lrLangDetectModel = pickle.load(lrLangDetectFile)
    lrLangDetectFile.close()

    text = " ".join(text.split())
    text = text.lower()
    text = re.sub(r"\d+","", text)
    text = text.translate(translate_table)
    pred = lrLangDetectModel.predict([text])
    prob = lrLangDetectModel.predict_proba([text])
    return pred[0]
    
```
